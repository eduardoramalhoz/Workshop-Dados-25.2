```python
import pandas as pd

df = pd.read_csv('academia_redfit.csv')

df['data_matricula'] = pd.to_datetime(df['data_matricula'], errors= 'coerce')
df['evolucao_PGC'] = df['primeiro_PGC'] - df['ultimo_PGC']
df['frequencia_semanal_treino'] = (
    pd.to_numeric(df['frequencia_semanal_treino'], errors='coerce')
      .astype('Int64')
)

df['tipo_atividade'] = df['tipo_atividade'].replace({
    'Natacao' : 'Natação',
    'Swimming' : 'Natação',
    'Soccer' : 'Futebol',
    'Fut' : 'Futebol'
})

df = df.dropna()

df.to_csv("dados_tratados.csv", index=False)

df

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dados_tratados.csv")

bar_counts = df["tipo_atividade"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(bar_counts.index, bar_counts.values)  # sem definir cores
plt.xlabel("Tipo de atividade")
plt.ylabel("Quantidade de alunos")
plt.title("Distribuição por atividade")
plt.tight_layout()
plt.show()
```

```python
pie_counts = df["estado"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(pie_counts.values, labels=pie_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Proporção por disposição física")
plt.tight_layout()
plt.show()
```
